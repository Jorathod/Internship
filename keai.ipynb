```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
paper_title=soup.find("h2",class_="h5 article-title")
paper_title
```

```python
paper_title.text.replace('\n','').replace('\r','')
```

```python
date=soup.find("p",class_="article-date")
date
```

```python
date.text
```

```python
authors=soup.find("p",class_="article-authors")
authors
```

```python
authors.text
```

```python
titles=[]
for i in soup.find_all("h2",class_="h5 article-title"):
    titles.append(i.text.strip())

titles
```

```python
date_=[]
for i in soup.find_all("p",class_="article-date"):
    date_.append(i.text)

date_
```

```python
authors_=[]
for i in soup.find_all("p",class_="article-authors"):
    authors_.append(i.text)

authors_
```

```python
import pandas as pd
df=pd.DataFrame({"Titles":titles,"Date":date_,"Authors":authors_})
df
```
