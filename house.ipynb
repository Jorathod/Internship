```python
soup=BeautifulSoup(page.content)
soup
```

```python
house=soup.find("a",class_="overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full")
print(house.text)
```

```python
houses=[]
for i in soup.find_all("a",class_="overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full"):
    houses.append(i.text)
    
houses
    
```

```python
location=soup.find("div",class_="mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0.1p po:max-w-95")
location
```

```python
loc=[]
for i in soup.find_all("div",class_="mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0.1p po:max-w-95"):
    loc.append(i.text)
loc
```

```python
a=soup.find("div",class_="flex",id="unitCode")
a
```

```python
area=[]
for i in soup.find_all("div",class_="flex",id="unitCode"):
    area.append(i.text)
area
```

```python
emi=soup.find("div",class_="flex", id="minimumRent")
emi.text
```

```python
EMi=[]
for i in soup.find_all("div",class_="flex", id="minimumRent"):
    EMi.append(i.text)
EMi
```

```python
p=soup.find("div",class_="font-semi-bold heading-6",id="roomType")
p.text
```

```python
price=[]
for i in soup.find_all("div",class_="font-semi-bold heading-6",id="roomType"):
    price.append(i.text)
price
```
